## Example3 -> scaler values

```python
import torch 

x = torch.tensor(6.7)  # Input feature
y = torch.tensor(0.0)  # True label (binary)

w = torch.tensor(1.0)  # Weight
b = torch.tensor(0.0)  # Bias
```

```python
#Binary Cross Entropy 
def BCE(prediction , target) :
    epsilon = 1e-8 #to prevent log(0)
    prediction = torch.clamp(prediction , epsilon , 1-epsilon) #torch.clamp-> keeps values withing the safe range , (x , min ,max) -. anything below min - make it min nd anything above max - make it max
    return -(target * torch.log(prediction) + (1-target) * torch.log(1-prediction))
```

```python
z =  w*x + b
y_pred = torch.sigmoid(z)
loss = BCE(y_pred, y)
```

```python
loss
```

```python
# Derivatives:
# 1. dL/d(y_pred):
dloss_dy_pred = (y_pred - y)/(y_pred*(1-y_pred))
# 2. dy_pred/dz:
dy_pred_dz = y_pred * (1 - y_pred)
# 3. dz/dw and dz/db: 
dz_dw = x 
dz_db = 1 

dL_dw = dloss_dy_pred * dy_pred_dz * dz_dw
dL_db = dloss_dy_pred * dy_pred_dz * dz_db
```

```python
print("Manula Gradients of Loss :")
print(f"w.r.t weight(dw)->{dL_dw}")
print(f"w.r.t weight(db)->{dL_db}")
```
